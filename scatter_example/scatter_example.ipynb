```python
%matplotlib inline
import numpy as np
import matplotlib.pylab as plt

# OK, this shit slows it way the fuck down.
from matplotlib import rc
rc('text', usetex=True)
```

```python
x, y = np.meshgrid(np.arange(20), np.arange(20))
```

```python
x = x.ravel()
y = y.ravel()
plt.plot(x,y, 'ko')
```

```python
# Make some random velocities
vel = np.random.rand(x.size) * 50
```

```python
fig, ax = plt.subplots()
# handy built-in colormaps shown here: https://matplotlib.org/examples/color/colormaps_reference.html
# Set value min and max to map the colors to
# use edgecolor to draw a black circle around each point
# s is for scale. 
cax = ax.scatter(x,y, c=vel, cmap='RdBu', vmax=30, vmin=10, edgecolors='k', s=20)
ax.set_xlabel(r'')
# make a colorbar
cb = fig.colorbar(cax)
cb.set_label(r'velocity (km s$^{-1}$)')
ax.set_xlabel('kpc')
fig.savefig('example.pdf')
```
