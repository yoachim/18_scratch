```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import healpy as hp

import lsst.sims.maf.db as db
import lsst.sims.maf.utils as utils
import lsst.sims.maf.metrics as metrics
import lsst.sims.maf.slicers as slicers
import lsst.sims.maf.stackers as stackers
import lsst.sims.maf.metricBundles as metricBundles
```

```python
dbFile = 'baseline_test10yrs.db'
# This connection checks which version it thinks the database is and should do the right thing schema-wise
conn = db.Database(dbFile, defaultTable='observations')
outDir='temp'
resultsDb = db.ResultsDb(outDir=outDir)
```

```python
sqls = ['filter="r" and night < 365',
        'filter="r" and night > 365 and night < 730',
        'filter="r" and night < 730']

bundleList = []
plotDict = {'percentileClip': 95.}
        
for sql in sqls:
    metric = metrics.CountMetric(col='mjd')
    slicer = slicers.HealpixSlicer(latCol='dec', lonCol='RA')
    bundleList.append(metricBundles.MetricBundle(metric,slicer,sql, plotDict=plotDict, runName='Baseline-ish'))
```

```python
bd = metricBundles.makeBundlesDictFromList(bundleList)
bg = metricBundles.MetricBundleGroup(bd, conn, outDir=outDir, resultsDb=resultsDb)
bg.runAll()
```

```python
bg.plotAll(closefigs=False)
```
