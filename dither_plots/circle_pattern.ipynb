```python
%matplotlib inline
import matplotlib.pylab as plt
import healpy as hp
import numpy as np
import lsst.sims.maf.db as db
import lsst.sims.maf.metrics as metrics
import lsst.sims.maf.slicers as slicers
import lsst.sims.maf.stackers as stackers
import lsst.sims.maf.plots as plots
import lsst.sims.maf.metricBundles as metricBundles
import lsst.sims.maf.utils as utils

from mpl_toolkits.basemap import Basemap
```

```python
# Let's see if there's a crazy circular pattern in the pointings
conn = db.Database('feature_baseline_10yrs.db', defaultTable='observations')
```

```python
data = conn.query_columns('observations', ['RA', 'dec'], '')
```

```python
plt.plot(data['RA'], data['dec'], 'bo', alpha=.1, markersize=.05)
```

```python
# Let's look at a spot seems smooth above.
x, y = utils.gnomonic_project_toxy(data['RA'], data['dec'], np.radians(100), 0)
plt.plot(x,y, 'bo', alpha=.1, markersize=.1)
plt.xlim([-.5,0.5])
plt.ylim([-.5,0.5])
```

```python
# now a spot that looks like it has an empty spot in the middle.
x, y = utils.gnomonic_project_toxy(data['RA'], data['dec'], np.radians(180), 0)
plt.plot(x,y, 'bo', alpha=.1, markersize=.1)
plt.xlim([-.5,0.5])
plt.ylim([-.5,0.5])
```
