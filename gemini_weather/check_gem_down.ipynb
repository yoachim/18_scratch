```python
import numpy as np
import matplotlib.pylab as plt
%matplotlib inline
```

```python
dt = np.load('2014_exp.npz')
```

```python
times = dt['exp_times'].copy()
dt.close()
```

```python
plt.plot(times/3600.)
plt.ylabel('Hours exposed')
plt.xlabel('night')
```

```python
no_obs = np.where(times == 0)[0]
print('number of nights no exposures = %i' % np.size(no_obs))
print('number of nights with exp = %i' % np.size(np.where(times > 0)[0]))
```

```python
plt.hist(times/3600., bins=20)
```
